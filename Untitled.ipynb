```python
from sklearn.datasets import load_digits
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectKBest, chi2
X = df2.drop("Diabetes_binary", axis = 1)
y = df2["Diabetes_binary"]

# Define the classifier
classifier = RandomForestClassifier()

# Define a range of k values to try
k_values = list(range(1,21))

# Store the results
results = {}

for k in k_values:
    print(k)
    # Create a feature selector with k features
    feature_selector = SelectKBest(chi2, k=k)
    X_selected = feature_selector.fit_transform(X, y)

    # Evaluate the model with cross-validation
    scores = cross_val_score(classifier, X_selected, y, cv=5, scoring='accuracy')

    # Store the average accuracy for this k value
    results[k] = scores.mean()

# Find the best k based on cross-validated performance
best_k = max(results, key=results.get)

print("Cross-validated accuracy for different k values:")
for k, accuracy in results.items():
    print(f"K={k}: {accuracy:.4f}")

print(f"\nBest k: {best_k} with accuracy {results[best_k]:.4f}")
```
